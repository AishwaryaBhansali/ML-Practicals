```python
from math import *

def distance_formula(longitude1, latitude1, longitude2, latitude2):
    #Haversine formula
    travel_dist = []

    for pos in range(len(longitude1)):
        lat1, long1, lat2, long2 = map(radians, [longitude1[pos], latitude1[pos], longitude2[pos], latitude2[pos]])
        dist_lon = lon2-lon1
        dist_lat = lat2 - lat1
        a = sin(dist_lat/2)**2 + cos(lat1) * cos(lat2) * sin(dist_lon/2)**2
        c = 2 * asin(sqrt(a)) * 6371 
        travel_dist.append(c)
    return travel_dist
```
