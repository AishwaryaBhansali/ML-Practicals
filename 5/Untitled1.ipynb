```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.array([
    [5, 2, 1, 0],
    [0, 1, 3, 1],
    [4, 0, 3, 5],
    [5, 1, 0, 3],
    [3, 2, 0, 1]
], dtype=float)
```

```python
print(R)
```

```python
U, sigma, Vt = np.linalg.svd(R, full_matrices=False)
```

```python
print(U)
```

```python
print(sigma)
```

```python
print(Vt)
```

```python
k=2
sigma_k = np.diag(sigma[:k])
```

```python
R_approx = np.dot(U[:, :k], np.dot(sigma_k, Vt[:k, :]))
```

```python
R_approx
```
