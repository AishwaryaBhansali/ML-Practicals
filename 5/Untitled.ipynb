```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.array([
    [5, 3, 0, 1],
    [4, 0, 0, 1],
    [1, 1, 0, 5],
    [1, 0, 0, 4],
    [0, 1, 5, 4],
], dtype=float)
```

```python
print(f"Original rating matrix with missing values : {R}")
```

```python
U, sigma, Vt = np.linalg.svd(R, full_matrices=False)
```

```python
print(f"User feature matrix : {U}")
```

```python
print(f"Singlular matrix : {sigma}")
```

```python
print(f"Item feature matrix : {U}")
```

```python
k = 2
sigma_k = np.diag(sigma[:k])
```

```python
R_approx = np.dot(U[:, :k], np.dot(sigma_k, Vt[:k, :]))
```

```python
print("Approx matrix : ")
print(np.round(R_approx, 2))
```

```python
fig, axes= plt.subplots(1, 2, figsize=(12,5))
axes[0].imshow(R, cmap="Blues")
axes[0].set_title("Original Rating Matrix")
axes[0].axis("off")

axes[1].imshow(R_approx, cmap="Blues")
axes[1].set_title(f"Approximated Matrix (Rank-{k})")
axes[1].axis("off")
plt.show()
```
