```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)

x = np.sort(np.random.rand(100)*10).reshape(-1,1)
y = np.sin(x).ravel() + np.random.normal(0, 0.2, x.shape[0])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
depths = range(1,15)
train_error = []
test_error = []
for d in depths:
    model = DecisionTreeRegressor(max_depth=d)
    model.fit(x_train, y_train)
    train_pred = model.predict(x_train)
    test_pred = model.predict(x_test)
    train_error.append(mean_squared_error(y_train, train_pred))
    test_error.append(mean_squared_error(y_test, test_pred))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(depths, train_error, label="Traning error", marker="o")
plt.plot(depths, test_error, label="Test error", marker="o")
plt.xlabel('Model Complexity (Tree Depth)')
plt.ylabel('Mean Squared Error')
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.grid(True)
plt.show()
```
